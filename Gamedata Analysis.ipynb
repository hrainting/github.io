```python
from itertools import combinations
print(list(combinations([1,2,3,4,5], 3)))
```

```python
def allist(i):
    lst_all = []
    if type(i)==str:
        tag_list_x = i.split(',')
        for j in range(len(tag_list_x)):   
            if j>=1:
                lst = combinations(tag_list_x, j+1)
                lst_all.append(list(lst))
    return str(lst_all)
```
