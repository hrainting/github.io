```python
#下载图片
def download_img(i,img_url):
#下载图片存储的路径
    path = r'D://paper//game//data//pictures//'
    r = requests.get(img_url,headers,stream=True)
    # print(r.status_code) # 返回状态码
    if r.status_code == 200:
        # 截取图片文件名
        img_name = str(i)+'.jpg'
        with open(path+'./'+str(img_name), 'wb') as f:
            f.write(r.content)
        return True
```
